## .hea files

```python
check_dict1 = dict()
for i in range(7):check_dict1[i] = set()
check_dict2 = dict()
for i in range(10):check_dict2[i] = set()
for dir in load_dict.keys():
    dir = dir.replace('mat', 'hea')
    heads = open(dir).readlines()
    first = heads[0].split(' ')
    for n, i in enumerate(first):
        check_dict1[n].add(i)
    second = heads[1].split(' ')
    for n, i in enumerate(second):
        check_dict2[n].add(i)
```

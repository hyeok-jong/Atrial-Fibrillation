```python
import numpy as np
import os
import matplotlib.pyplot as plt
import ecg_plot
from scipy.io import loadmat


# https://archive.physionet.org/physiobank/database/challenge/2017/
# https://physionet.org/content/challenge-2017/1.0.0/

```

```python
def load_challenge_data(filename):
    x = loadmat(filename)
    data = np.asarray(x['val'], dtype=np.float64)
    new_file = filename.replace('.mat','.hea')
    input_header_file = os.path.join(new_file)
    header_data=open(input_header_file,'r').readlines()
    return data, header_data
```

```python
filename = './data/training/A00/A00001.mat'
data, header_data = load_challenge_data(filename)
print(data.shape, header_data)

plt.figure(figsize = (20,3))
plt.subplot(2,1,1)
plt.plot(data.squeeze()[:5000])
plt.subplot(2,1,2)
plt.plot(data.squeeze()[-5000:])

```
